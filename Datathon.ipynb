```python
import pandas as pd
activity_df = pd.read_csv('https://raw.githubusercontent.com/Gaurang-1402/nyu-2022-datathon/main/activity_played.csv')
activity_df['occur_date'] = pd.to_datetime(activity_df['occur_date'])


def get_dataframe_window(dataframe, start_date, end_date):
  dataframe_window = dataframe.loc[activity_df['occur_date'] <= end_date]
  dataframe_window = dataframe_window.loc[dataframe_window['occur_date'] >= start_date]
  hacker_ids = [8018]
  dataframe_window = dataframe_window[dataframe_window.account_id.isin(hacker_ids) == False]
  return dataframe_window
```

```python
before_halloween_time = ('2020-10-22', '2020-10-28')
before_halloween = get_dataframe_window(activity_df, before_halloween_time[0], before_halloween_time[1])


print(before_halloween)
```

```python
during_halloween_time = ('2020-10-29', '2020-11-04')
during_halloween = get_dataframe_window(activity_df, during_halloween_time[0], during_halloween_time[1])
print(during_halloween)
```

```python
after_halloween_time = ('2020-11-05', '2020-11-11')
after_halloween = get_dataframe_window(activity_df, after_halloween_time[0], after_halloween_time[1])
print(after_halloween)
```

```python
from collections import defaultdict
def total_metric_per_week(week, metric):
  total_metric_per_day = defaultdict(int)
  unique_days = week['occur_date'].unique()
  for index, row in week.iterrows():
    day = row['occur_date']
    total_metric_per_day[day] += row[metric]
  return total_metric_per_day
```

```python
before_halloween_total_kills = total_metric_per_week(before_halloween, "kills")
print("Total number of kills per day in week before halloween event: ", before_halloween_total_kills)
print(sum(before_halloween_total_kills.values()))
```

```python
during_halloween_total_kills = total_metric_per_week(during_halloween, "kills")
print("Total number of kills per day in week during halloween event: ", during_halloween_total_kills)
print(sum(during_halloween_total_kills.values()))
```

```python
after_halloween_total_kills = total_metric_per_week(after_halloween, "kills")
print("Total number of kills per day in week after halloween event: ", after_halloween_total_kills)
print(sum(after_halloween_total_kills.values()))
```

```python
import matplotlib.pyplot as plt

def event_axes(before_metric, during_metric, after_metric):
  event_dates = []
  event_total_metric = []

  for event_timeframe in [before_metric, during_metric, after_metric]:
    for date, total_metric in sorted(event_timeframe.items()):
      event_dates.append(date)
      event_total_metric.append(total_metric)

  return event_dates, event_total_metric
```

```python
def plot_bar_charts(chart_title, event_type, before, during, after, before_event_time, during_event_time, after_event_time):
  halloween_dates, halloween_total_metric = event_axes(before, during, after)

  # print(halloween_total_metric)
  before_arr = [metric for date, metric in sorted(before.items())]
  during_arr = [metric for date, metric in sorted(during.items())]
  after_arr = [metric for date, metric in sorted(after.items())]

  # print(before_arr)
  # print(during_arr)
  # print(after_arr)

  N = 3
  ind = np.arange(N) 
  width = 0.1
  plt.figure(figsize = (21, 5))
  day1 = [before_arr[0], during_arr[0], after_arr[0]]
  day2 = [before_arr[1], during_arr[1], after_arr[1]]
  day3 = [before_arr[2], during_arr[2], after_arr[2]]
  day4 = [before_arr[3], during_arr[3], after_arr[3]]
  day5 = [before_arr[4], during_arr[4], after_arr[4]]
  day6 = [before_arr[5], during_arr[5], after_arr[5]]
  day7 = [before_arr[6], during_arr[6], after_arr[6]]


  xvals = day1
  bar1 = plt.bar(ind, xvals, width, color = 'r')
    
  yvals = day2
  bar2 = plt.bar(ind+width, yvals, width, color='g')
    
  zvals = day3
  bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')

  avals = day4
  bar4 = plt.bar(ind+width*3, avals, width, color = 'c')

  bvals = day5
  bar5 = plt.bar(ind+width*4, bvals, width, color = 'm')

  cvals = day6
  bar6 = plt.bar(ind+width*5, cvals, width, color = 'y')

  dvals = day7
  bar7 = plt.bar(ind+width*6, dvals, width, color = 'k')
    

  plt.xlabel("Days")
  plt.ylabel('Total {title}'.format(title=chart_title))
  plt.title("Total {title} Each Day Before, During, and After {event} Event".format(title=chart_title, event=event_type))
    
  xaxis_labels = ['Before ({0}-{1})'.format(before_event_time[0],before_event_time[1]), 'During ({0}-{1})'.format(during_event_time[0],during_event_time[1]), 'After ({0}-{1})'.format(after_event_time[0],after_event_time[1])]
  plt.xticks(ind+width*3,xaxis_labels)
  plt.show()
```

```python
def plot_line_charts(metric, event, before, during, after, before_event_time, during_event_time, after_event_time):
  b = sorted(before.items())
  d = sorted(during.items())
  a = sorted(after.items())
  timedays = [ts.strftime('%Y-%m-%d') for ts, metric in b] + [ts.strftime('%Y-%m-%d') for ts, metric in d] + [ts.strftime('%Y-%m-%d') for ts, metric in a]
  timearr = [metric for ts, metric in b] + [metric for ts, metric in d] + [metric for ts, metric in a]
  plt.figure(figsize=(26,5))
  plt.plot(timedays, timearr)
  plt.title('Total {metric} Each Day for Time Around {event} Event'.format(metric=metric, event=event))
  plt.xlabel('Day')
  plt.ylabel('{metric}'.format(metric=metric))
  plt.show()
```

```python
plot_bar_charts("Kills", "Halloween", before_halloween_total_kills, during_halloween_total_kills, after_halloween_total_kills, before_halloween_time, during_halloween_time, after_halloween_time)
```
